```python
## Listing a count of the most number of players drafted in the nba by college
```

```python
Schools_df =df1['school'].value_counts().reset_index().sort_values(by='count', ascending =False).nlargest(n=10, columns = 'count')
```

```python
Schools_df =df1['school'].value_counts().reset_index().sort_values(by='count', ascending =False).nlargest(n=10, columns = 'count')
print(Schools_df)
```

```python
sns.barplot(data= Schools_df, x= 'school', y= 'count')
plt.xticks(rotation=45)  # Rotate x-axis labels 45 degrees
plt.tight_layout()       # Prevent labels from getting cut off
plt.show()
```

```python
## Looking at players in the NBA from The University of Tennessee Knoxville and dropping null values from the draft_number column--
```

```python
Tenn_df = df1[df1['school']=='Tennessee'].sort_values('draft_year', ascending =False)
```

```python
Tenn_df[Tenn_df['draft_number'].notna()]
```

```python
## Adding a years_played column by looking at the start and end year in the NBA
```

```python
Tenn_df['years_played'] = Tenn_df['to_year'] - Tenn_df['from_year'] + 1
Tenn_df.sort_values('years_played', ascending =False)
```

```python
Tenn_df.rename(columns={'person_id': 'player_id'}, inplace=True)
```

```python
Years_played = Tenn_df[['display_first_last','years_played']].sort_values('years_played',ascending = False).nlargest(n=10, columns = 'years_played')
```

```python
Tenn_df[['player_id','display_first_last']]
```

```python
sns.barplot(data= Years_played, x= 'display_first_last', y= 'years_played')
plt.xticks(rotation=45)  # Rotate x-axis labels 45 degrees
plt.tight_layout()       # Prevent labels from getting cut off
plt.show()
```
